```python
import pandas as pd
```

Load the CSV file


```python
df = pd.read_csv('/content/data.csv')
```

Display the initial DataFrame

```python
print("Initial DataFrame:")
print(df)
```

**1. Handle missing values**

Fill missing 'Age' with the mean age

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

Fill missing 'Salary' with the median salary

```python
df['Salary'].fillna(df['Salary'].median(), inplace=True)
```

Drop rows with missing values in the 'Department' column

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
```

```python
df.dropna(subset=['Department'], inplace=True)
```

**2. Perform other transformations**

Convert 'Salary' to integer type

```python
df['Salary'] = df['Salary'].astype(int)
```

Rename the 'Department' column to 'Dept

```python
df.rename(columns={'Department': 'Dept'}, inplace=True)
```

Display the transformed DataFrame

```python
print("\nTransformed DataFrame:")
print(df)
```

Dropping duplicate rows based on the 'Name' column

```python
 df_no_duplicates = df.drop_duplicates(subset='Name')
```

Display the DataFrame after removing duplicates

```python
df_no_duplicates.head()
```

Dropping rows with more than 50% missing values (in this case more than 2 out of 4)

```python
df_filtered_missing = df.dropna(thresh=2)
```
